```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Tanmay49/SEM7/main/ML/sales_data_sample.csv',encoding = 'unicode_escape')
df.head()
```

```python
df = df[['QUANTITYORDERED','PRICEEACH','SALES']]
```

```python
from sklearn.cluster import KMeans
lis = []
for k in range(1,15):
  kmeans = KMeans(n_clusters = k)
  kmeans.fit(df)
  lis.append(kmeans.inertia_)
```

```python
import seaborn as sns
```

```python
sns.lineplot(x = range(1,15),y = lis)
```

```python
# 4 is elbow as after that point no much change in value of
kmeans = KMeans(n_clusters = 4)
df['Cluster No.'] = kmeans.fit_predict(df)
df
```
